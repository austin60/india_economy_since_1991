## World Bank Data of Indian Economy since 1991

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### read file 

```python
india_data=pd.read_csv('World_Bank_Data_India.csv')
india_data.head()
```

### drop multiple null columns

```python
india_data=india_data.drop(['LF_ADVN_EDU','LF_BASC_EDU','LF_INTM_EDU','XPD_TOTL_EDU','D_COM','D_INJ','D_NC'], axis=1)
india_data.head()
```

### Qn1. What is the performance of GDP in the two sectors

```python
year=[year for year,df in india_data.groupby('Years')]
IND=india_data['GDP_IND']
AGR=india_data['GDP_AGR']

plt.plot(year, IND, label='Industry', marker='o')
plt.plot(year, AGR, label='Agric', marker='o')
plt.legend(loc="lower left")

plt.xlabel("year")
plt.ylabel("GDP index")
plt.title('Agriculture and Industry Employment GDP ')
plt.xticks(year, rotation ='vertical')
plt.show()
```

### Qn2. Employment performance

```python
employment=india_data['EMP_TOTL']
unemployment=india_data['UEM_TOTL']
year=[year for year,df in india_data.groupby('Years')]

fig, ax1=plt.subplots()

ax2=ax1.twinx()
ax1.bar(year,employment)
ax2.plot(year,unemployment,'r-',label='unemployment')

ax1.set_xlabel('year')
ax1.set_ylabel("Employment in millions")
ax2.set_ylabel('unemployment')
ax1.set_xticklabels(year,rotation='vertical',size=8)
plt.title('Employment against Unemployment since 1991 ')
plt.xticks(year, rotation ='vertical')
plt.legend(loc="lower left")
plt.show()
```

### Qn3. Age groups ( Labour force ) population

```python
year=[year for year,df in india_data.groupby('Years')]
under14=india_data['POP_014']
over14=india_data['POP_1564']
over65=india_data['POP_65']
DR=india_data['DR']

plt.plot(year,under14, label='under 15',color='blue')
plt.plot(year,over14, label='15 to 64',color='green')
plt.plot(year,over65, label='above 64',color='orange')

plt.xlabel("year")
plt.ylabel("population in millions")
plt.legend(loc='upper left')
plt.show()
#india_data.head()
```

### Qn4. What are the factors  

```python
year=[year for year,df in india_data.groupby('Years')]
infertility=india_data['INFL']
fertility=india_data['FRTL']
birth_rate=india_data['BR']
death_rate=india_data['DR']

plt.plot(year,infertility, color='orange',label='infertility')
plt.plot(year,fertility, color='blue',label='fertility')
plt.plot(year,birth_rate, color='green',label='birth rate')
plt.plot(year,death_rate, color='red',label='death rate')
plt.title("Population factors")
plt.xticks(year,rotation='vertical')
plt.ylabel("Rate in percentage")
plt.legend()
plt.grid()

plt.show()
```
